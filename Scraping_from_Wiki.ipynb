**Scraping Data from Wikipedia (Pandas, BeautifulSoup)**

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Finland'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
table = soup.find('table')
```

```python
titles = table.find_all('th')
```

```python
print(titles)
```

```python
table_titles = [title.text.strip() for title in titles]
print(table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = table_titles)
df
```

```python
columns_data = table.find_all('tr')
```

```python
for row in columns_data[1:]:
    row_data = row.find_all('td')
    table_data = [data.text.strip() for data in row_data]
    print(table_data)
    
    length = len(df)
    df.loc[length]=table_data
```
