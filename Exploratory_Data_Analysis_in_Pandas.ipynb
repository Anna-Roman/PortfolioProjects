**Exploratory data analysis**  
Exploring world population data in Pandas  
Data: https://github.com/AlexTheAnalyst/PandasYouTubeSeries/blob/main/world_population.csv 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read data
df =  pd.read_csv(r"C:\Users\anetr\OneDrive\Documents\world_population.csv")
df.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# view some meta data about our data
df.info()
```

```python
#looking at min, max and mean of our data
df.describe()
```

```python
# how many nulls do we have and in which columns
df.isnull().sum()
```

```python
# unique values
df.nunique()
```

```python
#sorting data by '2022 Population' column in descending order
df.sort_values(by='2022 Population', ascending=False).head(10)
```

```python
# checking if there is any correlation in the data
df.corr(numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.rcParams['figure.figsize'] = (20, 10)
plt.show()
```

```python
#grouping data by 'Continent' and sorting in descending order, , looking at the means
df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

```python
df.columns
```

Grouping data by 'Continent' and sorting in descending order, looking at the means, and choosing columns only with populaition. Then transposin resulting data frame.

```python
df2 = df.groupby('Continent')[df.columns[5:13]].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df2.transpose()
```

```python
df3 = df2.transpose()
#visualising our data
df3[df3.columns[::-1]].plot()
```
